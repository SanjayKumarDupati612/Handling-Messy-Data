```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

df
```

```python
df.info()
```

```python
impute_values = {
    'age': round(df['age'].mean(), 0),
    'city': df['city'].mode()[0]
}
df.fillna(impute_values, inplace=True)
df["age"] = round(df["age"]).astype(int)
df
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df = pd.get_dummies(df, columns=['city'])
df = pd.DataFrame(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["gender_encoded"] = le.fit_transform(df["gender"])
df.drop('gender', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
standard_copy = df.copy()

standard_copy[['age', 'annual_income']] = scaler.fit_transform(standard_copy[['age', 'annual_income']])
standard_copy
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
```

```python
minmax_copy = df.copy()
minmax_copy[['age', 'annual_income']] = minmax_scaler.fit_transform(minmax_copy[['age', 'annual_income']])
minmax_copy
```

Standardization scales the data to have a mean of 0 and standard deviation of 1. It preserves the distribution of the data and very robust in handling outliers. Preferred to use when the algorithm is distance-based, data have outliers, and there is no requirement for range.

MinMax scaling scales the data between the range [0, 1]. It doesn't preserve the distribution of the data. Use cases include Linear/Statistical model, few/no outliers, when the scaled values need to be in a specific range.
